# Hospital Management

```python
class hospital:
    def __init__(self,room,doc):
        self.ward=room
        self.doctor=doc
        self.stat=[]
    def ward_availablity(self):
      print("available wards are :",','.join(self.ward))
    def status(self):
      print("patient is in",''.join(self.stat))
    def doctor_available(self,dr):
      if dr in self.doctor:
        self.doctor.remove(dr)
        print("doctor available pls go to",dr," ward")
      else:
        print("doctor not available")
    def ward_available(self,wd):
      if wd in self.ward:
        self.ward.remove(wd)
        self.stat.append(wd)
        print("patient amitted in",wd)
      else:
        print("sorry now ward not available")
    def change_wards(self,cw):
      if cw in self.ward:
        self.ward.remove(cw)
        self.ward.append(','.join(self.stat))
        self.stat.pop(0)
        self.stat.append(cw)
        print("patient changed to",cw)
      else:
        print("sorry ward not available")
    def dis(self):
      self.ward.append(','.join(self.stat))
      print("patient discharged")


class doctors():
    def war(self):
        print("where need to admit the patient")
        print("\n icu,\n emergency \n general")
        self.wa=input()
        return self.wa
    def change(self):
        print("change patient ward to:")
        self.chang=input()
        return self.chang
    
        

class patient():
    def form(self):
        print("name: ")
        self.name=input()
        print("need to meet: eye doctor, hear doctor, nose doctor ")
        self.do=input()
        return self.do


   
```

```python
hos= hospital(room=["icu","icu","emergency","general","general"],doc=["eye","eye","nose","hear"])
pat=patient()
docto=doctors()
```

```python
while True:
    print("***********srs hospitals************")
    print("\n ARE YOU")
    print("\n 1-PATIENT \n 2-DOCTOR")
    option=int(input())
    if option==1:
      requestdoctor=pat.form()
      hos.doctor_available(requestdoctor)

    elif option==2:
      print("\n 3-check availabilty of ward \n 4-admit patient \n 5-check status of patient \n 6-change the ward \n 7-discharge the patient")
      opt=int(input())
      if opt==3:
        hos.ward_availablity()
      elif opt==4:  
        reqward=docto.war()
        hos.ward_available(reqward)
      elif opt==5:
        hos.status()
      elif opt==6:
        changeward=docto.change()
        hos.change_wards(changeward)
      elif opt==7:
        hos.dis()
      else:
        print("INVALID REQUEST")
    else:
      print("INVALID REQUEST")

```
